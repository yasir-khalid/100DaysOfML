```python
from data_import import GDP, life_expectancy, population
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
plt.plot(GDP, life_expectancy)
plt.xscale('log')
```

```python
plt.xscale('log') 
plt.xlabel("GDP per Capita")
plt.ylabel("Life Expectancy")
plt.title("World Development chart")

# Definition of tick_val and tick_lab
tick_val = [1000,10000,100000]
tick_lab = ['1k','10k','100k']

# Adapt the ticks on the x-axis using plt.xticks([0,1,2], ["one","two","three"])
plt.xticks(tick_val, tick_lab)

# Adding bubble size according to a third variable
# Wouldn't it be nice if the size of the dots corresponds to the population?
np_population = np.array(population)
# Created a numpy array to double each element to enhance affect of bubble
plt.scatter(GDP, life_expectancy, s = np_population)
```
