```python
import seaborn as sns
```

```python
# seaborn has some pre-added datasets for practise
tips = sns.load_dataset("tips")
flights = sns.load_dataset("flights")
%matplotlib inline
tips.head()
```

```python
sns.rugplot(tips["total_bill"]) #you'll get to see around what values data is congested
# each dash in rugplot has it's normal distribution and when all those distributions are superimposed you get
# KDE (Kernel density estimation)
```

```python
# used to visualise univariate data i.e. one variable distribution
sns.distplot(tips["total_bill"])
```

```python
sns.distplot(tips["total_bill"], kde = False, bins = 50)
```

```python
# used to visualise bivariate data i.e. compare two variable distribution
sns.jointplot(x = "total_bill",y = "tip",data = tips)
```

```python
sns.jointplot(x = "total_bill",y = "tip",data = tips, kind = "hex")
```

```python
sns.jointplot(x = "total_bill",y = "tip",data = tips, kind = "reg")
```

```python
sns.jointplot(x = "total_bill",y = "tip",data = tips, kind = "kde")
```

```python
sns.pairplot(tips) #does the same thing as joint plot for all possible commbination of columns
```

```python
sns.pairplot(tips, hue = "sex") # hue differentiates the datapoints based off the categorical column
```

# Categorical data visualisation

```python
sns.countplot(x = "sex",data = tips)
```

```python
sns.barplot(x = "sex",y = "total_bill", data = tips) #A bit advanced than count plot
# rather than having just counts on y-axis you can have anything using 'estimator' argument, default: mean
```

```python
# compare quantative data against categorical
sns.boxplot(x = "day", y = "total_bill", data = tips)
```

```python
sns.boxplot(x = "day", y = "total_bill", data = tips, hue = "smoker")
# hue argument adds another category on top of the graph against which y-axis was being measured
```
